```python
#Create a visual showing the relationship between horsepower and fuel economy. 
import matplotlib.pyplot as plt
mpg.head()
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.title('Relationship between horsepower and fuel economy')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon')
plt.show
```

```python
#Create a table summarizing fuel economy by country of origin
import seaborn as sns
import pandas as pd

print(mpg.head())

mpg_summary = mpg.groupby('origin')['mpg'].mean().reset_index()

mpg_summary.columns = ['Country of Origin', 'Average MPG']

print(mpg_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
mpg = sns.load_dataset("mpg")

mpg_trend = mpg.groupby('model_year')['mpg'].mean().reset_index()

sns.lineplot(x='model_year', y='mpg', data=mpg_trend, marker='o')

plt.title('Average Fuel Economy (MPG) Over Time')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "mpg" dataset
mpg = sns.load_dataset('mpg')

# Create a histogram of MPG
sns.histplot(mpg['mpg'], bins=10, kde=True)

# Add a title and labels
plt.title('Distribution of Fuel Economy (MPG)')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "mpg" dataset
mpg = sns.load_dataset('mpg')

# Create a bar plot of average MPG by number of cylinders
sns.barplot(x='cylinders', y='mpg', data=mpg, errorbar=None)

# Add a title and labels
plt.title('Average Fuel Economy (MPG) by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Miles per Gallon (MPG)')

# Show the plot
plt.show()
```
